```python
import numpy as np
```

```python
results = np.load('./output/quadratic_64_result.npy')
```

```python
import matplotlib.pyplot as plt

for epoch in range(results.shape[0]):
    epoch_results = results[epoch]
    x = epoch_results[:, 1]
    y_true = epoch_results[:, 2]
    y_pred = epoch_results[:, 3]

    sorted_x = x[x.argsort()]
    sorted_y_true = y_true[x.argsort()]
    sorted_y_pred = y_pred[x.argsort()]

    plt.plot(sorted_x, sorted_y_true)
    plt.plot(sorted_x, sorted_y_pred)
    plt.title(f"Epoch {epoch}")
    plt.show()
```
